```python
import requests
import re
import csv
import pymysql

class TiebaSpider(object):
    def __init__(self,tiebaName):
        self.base_url = 'https://tieba.baidu.com/f?kw='+tiebaName+'&ie=utf-8&pn={}'
        self.headers = {
            "User-Agent":'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.131 Safari/537.36'
        }
    
    #构造请求链接:
    def get_url_list(self):
        url_list = []
        for i in range(1):
            url_list.append(self.base_url.format(i*50))
        return url_list
    
    #请求界面的方法:
    def get_page_info(self,url):
        #所有的html数据
        reponse = requests.get(url=url,headers=self.headers)
        
        #调用解析的方法
        return self.parse_page_info(reponse.content.decode("utf-8")) 
    
     #解析界面的方法
    def parse_page_info(self,html):
        pattern = re.compile('<li class=" j_thread_list clearfix".*?<a rel="noreferrer".*?href="(.*?)".*?title="(.*?)".*?</a>',re.S)
    
        return re.findall(pattern,html)
    
     #保存到本地(txt)的方法
    def save_to_txt(self,info):
         for tuple_value in info:
            in_fo_str = '帖子的信息:'+tuple_value[1]+'帖子的链接:https://tieba.baidu.com'+tuple_value[0]+'\n'
            #mac 默认是utf -8 linux
            with open("./tieba.txt","a")as f:
                f.write(in_fo_str)
    
     #保存到本地(csv)
    def save_to_csv(self,info):
        for tupel_value in info:
            #newline:excle 默认\r\n
            #encoding 解决中文乱码   
            with open('./tieba.csv','a',newline='',encoding='utf-8-sig')as f: 
                writer = csv.writer(f)
                #列表
                writer.writerow(list(tupel_value))
    
    #保存到mysql
    def sava_data_Base(self,info):
         #连接数据库 127.0.0.1
        conn = pymysql.connect(host='localhost',user='root',passwd='9483624ljk',db='tieba',port=3306,charset='utf8mb4')
         #游标对象(操纵数据库必须使用游标提供了很多操作数据库的方法),比如执行sql
        cursor = conn.cursor()
        for tupel_value in info:
            #执行sql语句
            sql = "insert in to tie_ba(info,url)values('{}','{}')".format(tuple_value[0],tuple_value[1])
            #执行sql语句
            cursor.execute(sql)
            #提交事务(查询可以不用使用)
            conn.commit()
         #先关闭游标
        cursor.close()
        #关闭数据库
        conn.close()
    
     #所有的爬虫逻辑都放在此方法里面
    #self
    def get_spider(self):
        #先获取请求连接
        url_list = self.get_url_list()
        #发送请求
        for url in url_list:
            #完整的html
            #解析完毕的数据 (url,title)
            #请求数据并且解析
            info = self.get_page_info(url)
            print(info)
            #txt格式
            self.save_to_csv(info)
            #保存csv格式
            #1 
            #2
            #保存数据
    
tiebaspider = TiebaSpider("lol")
tiebaspider.get_spider()
    
```
